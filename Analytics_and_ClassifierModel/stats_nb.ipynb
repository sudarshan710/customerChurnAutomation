```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cust_data.csv')
df.shape
```

```python
total_monthly_contract = df[df['Contract Period'] == 1].shape[0]
total_annual_contract = df[df['Contract Period'] == 12].shape[0]
total_binnual_contract = df[df['Contract Period'] == 24].shape[0]
print(total_monthly_contract, total_annual_contract, total_binnual_contract, total_monthly_contract+total_annual_contract+total_binnual_contract)

total_male = df[df['Gender'] == 'Male'].shape[0]
total_female = df[df['Gender'] == 'Female'].shape[0]
print(total_male, total_female, total_male+total_female)

total_senior_citizen = df[df['Senior Citizen'] == 'Yes'].shape[0]
sd = df[df['Senior Citizen'] == 'No'].shape[0]
print(total_senior_citizen, sd, total_senior_citizen+sd)

total_with_partner = df[df['Partner'] == 'Yes'].shape[0]
dd = df[df['Partner'] == 'No'].shape[0]
print(total_with_partner, dd, total_with_partner+dd)

total_with_dependents = df[df['Dependents'] == 'Yes'].shape[0]
ed = df[df['Dependents'] == 'No'].shape[0]
print(total_with_dependents, ed, total_with_dependents+ed)

more_than_year_tenure = df[df['Tenure Months'] >= 12].shape[0]
joined_this_quarter = df[df['Tenure Months'] <= 3].shape[0]
max_tenure = df['Tenure Months'].max()
min_tenure = df['Tenure Months'].min()
avg_tenure = df['Tenure Months'].mean()
print(more_than_year_tenure, joined_this_quarter, max_tenure, min_tenure, avg_tenure)

total_with_phone_service = df[df['Phone Service'] == 'Yes'].shape[0]
total_without_phone_service = df[df['Phone Service'] == 'No'].shape[0] 
print(total_with_phone_service, total_without_phone_service, total_with_phone_service+total_without_phone_service)

# Customers without phone service cannot have multiple lines

total_with_multiple_lines = df[df['Multiple Lines'] == 'Yes'].shape[0]
total_without_multiple_lines = df[df['Multiple Lines'] == 'No'].shape[0]
print(total_with_multiple_lines, total_without_multiple_lines, total_with_multiple_lines+total_without_multiple_lines)

total_with_DSL = df[df['Internet Service'] == 'DSL'].shape[0]
total_with_Fiber_optic = df[df['Internet Service'] == 'Fiber optic'].shape[0]
total_without_internet_service = df[df['Internet Service'] == 'No'].shape[0]
print(total_with_DSL, total_with_Fiber_optic, total_without_internet_service, total_with_DSL+total_with_Fiber_optic+total_without_internet_service)

# Customers without internet service cannot have online security or online backup or 
# device protection or tech support or streaming tv or streaming movies

totral_with_online_security = df[df['Online Security'] == 'Yes'].shape[0]
total_without_online_security = df[df['Online Security'] == 'No'].shape[0]
print(totral_with_online_security, total_without_online_security, totral_with_online_security+total_without_online_security)

total_with_online_backup = df[df['Online Backup'] == 'Yes'].shape[0]
total_without_online_backup = df[df['Online Backup'] == 'No'].shape[0]
print(total_with_online_backup, total_without_online_backup, total_with_online_backup+total_without_online_backup)

total_with_device_protection = df[df['Device Protection'] == 'Yes'].shape[0]
total_without_device_protection = df[df['Device Protection'] == 'No'].shape[0]
print(total_with_device_protection, total_without_device_protection, total_with_device_protection+total_without_device_protection)

total_with_tech_support = df[df['Tech Support'] == 'Yes'].shape[0]
total_without_tech_support = df[df['Tech Support'] == 'No'].shape[0]
print(total_with_tech_support, total_without_tech_support, total_with_tech_support+total_without_tech_support)

total_with_streaming_tv = df[df['Streaming TV'] == 'Yes'].shape[0]  
total_without_streaming_tv = df[df['Streaming TV'] == 'No'].shape[0]
print(total_with_streaming_tv, total_without_streaming_tv, total_with_streaming_tv+total_without_streaming_tv)

total_with_streaming_movies = df[df['Streaming Movies'] == 'Yes'].shape[0]
total_without_streaming_movies = df[df['Streaming Movies'] == 'No'].shape[0]
print(total_with_streaming_movies, total_without_streaming_movies, total_with_streaming_movies+total_without_streaming_movies)

total_paperless_billing = df[df['Paperless Billing'] == 'Yes'].shape[0]
total_no_paperless_billing = df[df['Paperless Billing'] == 'No'].shape[0]
print(total_paperless_billing, total_no_paperless_billing, total_paperless_billing+total_no_paperless_billing)

total_with_bank_transfer = df[df['Payment Method'] == 'Bank transfer (automatic)'].shape[0]
total_with_mailed_check = df[df['Payment Method'] == 'Mailed check'].shape[0]
total_with_credit_card = df[df['Payment Method'] == 'Credit card (automatic)'].shape[0]
total_with_electronic_check = df[df['Payment Method'] == 'Electronic check'].shape[0]
print(total_with_bank_transfer, total_with_mailed_check, total_with_credit_card, total_with_electronic_check, total_with_bank_transfer+total_with_mailed_check+total_with_credit_card+total_with_electronic_check)

average_monthly_charges = df['Monthly Charges'].mean()
average_total_charges = df['Total Charges'].mean()
print(average_monthly_charges, average_total_charges)

total_churn_yes = df[df['Churn Label'] == 'Yes'].shape[0]
total_churn_no = df[df['Churn Label'] == 'No'].shape[0]
print(total_churn_yes, total_churn_no, total_churn_yes+total_churn_no)
total_churn_yes_monthly = df[(df['Churn Label'] == 'Yes') & (df['Contract Period'] == 1)].shape[0]
total_churn_yes_annual = df[(df['Churn Label'] == 'Yes') & (df['Contract Period'] == 12)].shape[0]
total_churn_yes_binnual = df[(df['Churn Label'] == 'Yes') & (df['Contract Period'] == 24)].shape[0]
total_churn_no_monthly = df[(df['Churn Label'] == 'No') & (df['Contract Period'] == 1)].shape[0]
total_churn_no_annual = df[(df['Churn Label'] == 'No') & (df['Contract Period'] == 12)].shape[0]
total_churn_no_binnual = df[(df['Churn Label'] == 'No') & (df['Contract Period'] == 24)].shape[0]


max_CLTV_score = df['CLTV'].max()
min_CLTV_score = df['CLTV'].min()
average_CLTV_score = df['CLTV'].mean()
print(max_CLTV_score, min_CLTV_score, average_CLTV_score)

max_churn_score = df['Churn Score'].max()
min_churn_score = df['Churn Score'].min()
average_churn_score = df['Churn Score'].mean()
print(max_churn_score, min_churn_score, average_churn_score)
```
